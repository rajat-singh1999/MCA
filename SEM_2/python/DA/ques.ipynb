```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('CO2 Emissions_Canada.csv')
data = data.sort_values('CO2 Emissions(g/km)')
# Extract the relevant columns from the dataset
make = data['Make']
model = data['Model']
consumption = data['Fuel Consumption City (L/100 km)']
co2_emissions = data['CO2 Emissions(g/km)']
```

```python
plt.figure(figsize=(12, 6))
u_make = make.unique()
u_co2_em = co2_emissions.unique()
plt.bar(u_make[-10:], u_co2_em[-10:])
plt.xlabel('Make')
plt.ylabel('CO2 Emissions (g/km)')
plt.title('Top 10 Makes by CO2 Emissions')
plt.xticks(rotation=45)
plt.show()
```

```python
e_size = data['Cylinders']
plt.figure(figsize=(12, 6))
plt.plot(u_make[-10:], u_co2_em[-10:])
plt.xlabel('Maker')
plt.ylabel('CO2 Emissions (g/km)')
plt.title('CO2 Emissions vs Maker')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.hist(co2_emissions,edgecolor="black")
plt.xlabel('CO2 Emissions (g/km)')
plt.ylabel('Frequency')
plt.title('CO2 Emissions Distribution')
plt.show()
```

```python
make_counts = make.value_counts().head(5)
plt.figure(figsize=(8, 8))
plt.pie(make_counts, labels=make_counts.index, autopct='%1.1f%%')
plt.title('Top 5 Makes Contribution to CO2 Emissions')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.scatter(u_make[-10:], u_co2_em[-10:], alpha=0.5)
plt.xlabel('Maker')
plt.ylabel('CO2 Emissions (g/km)')
plt.title('CO2 Emissions vs Maker')
plt.show()
```
